```python
from math import sin, cos, sqrt, atan2, radians 
```

```python
data1 = pd.read_csv("C:/Users/renz/Downloads/Query_Executor_out.csv")
data2 = pd.read_csv("C:/Users/renz/Downloads/PORT4.csv")
```

```python
L2 = []
for i,j,k in zip(data1['MMSI2'], data1['LAT'], data1['LON']):
    L1 = []
    for l1, l2 in zip(data2['LATITUDE'], data2['LONGITUDE']):
        R = 6373.0 
        lat1 = radians(j) 
        lon1 = radians(k) 
        lat2 = radians(l1) 
        lon2 = radians(l2) 
        dlon = lon2 - lon1 
        dlat = lat2 - lat1 
        a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2 
        c = 2 * atan2(sqrt(a), sqrt(1 - a)) 
        distance = R * c
        L1.append(distance)
    L2.append(min(L1))
```

```python
data1['distance'] = L2
```

```python
data6 = pd.read_csv("C:/Users/renz/Downloads/DATA/BUSAN_LONBEACH_out.csv")
```

```python
# del data6['color']
del data6['MMSI']
del data6['LON']
del data6['ETA2']
# data6['ETA'] = data6['ETA'].apply(lambda x:x.replace('-', ''))
```

```python
import arrow
```

```python
data6_diff_days = data6.groupby('MMSI2')['ETA'].max().apply(lambda x: arrow.get(x))-data6.groupby('MMSI2')['ETA'].min().apply(lambda x: arrow.get(x))
```

```python
L=list()
for ix in range(data6.shape[0]):
    for i in data6_diff_days.index:
        if data6.loc[ix, 'MMSI2'] == i:
            L.append(data6_diff_days[i])
            break
```

```python
data6['days'] = L
```

```python
data6['ETA'] = data6['ETA'].apply(lambda x: x.replace('-', ''))
```

```python
data6['days'] = data6['days'].apply(lambda x: str(x)[:2])
```

```python
import numpy as np
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data6.iloc[:,:8], list(data6['days']), test_size=0.33, random_state=1)
```

```python
regr = LinearRegression()
```

```python
regr.fit(X_train, y_train)
```

```python
y_pred = regr.predict(X_test)
```

```python
y_sqrt = []
for i,j in zip(y_pred, y_test):
    y_sqrt.append((float(i)-float(j))**2)
```

```python
np.sqrt(np.mean(y_sqrt))
```
